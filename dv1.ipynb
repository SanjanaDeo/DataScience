```python
import pandas as pd
import seaborn as sns # Importing seaborn
sns.set_theme() # Setting the default seaborn plotting theme
```

```python
df = pd.read_csv("titanic.csv") # Reading the csv file and storing it as a Pandas dataframe in "df"
```

```python
df.head()
```

```python
sns.histplot(data=df, x=df['Fare']) # Plotting a histogram for x = Fare prices, as a histogram plot.
```

```python
sns.histplot(data=df, x=df['Fare'], kde=True, bins=10) # Setting Kernel Density Estimation = True for smoothening and getting a curve line for the plot.
# Also setting bins = 10, also known as the "intervals" or "grouping data into equal width"
```

```python
sns.scatterplot(data=df, x=df['Fare'], y=df['Age'])
# Defining a scatterplot to understand Fare vs Age factors. AKA: What person with age "X" paid "Y" price.
```

```python
sns.lineplot(data=df, x=df['2urvived'], y=df['Age'])
```

```python
sns.barplot(data=df, x="Embarked", y="2urvived")
```
